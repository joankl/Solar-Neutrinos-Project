# Notebook dedicated to observed the (remote preprocesed) MC oscillated data through Rampling Rejection

```python
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sn
```

# Load Data

```python
# Directories Definition
main_dir = '/home/joankl/data/solars/mc/bisMSB/'

b8__nonosc_nue_dir = 'b8Nue/non_osc/np_array_resume/'
b8__osc_nue_dir = 'b8Nue/osc/np_array_resume/'

b8__nonosc_numu_dir = 'b8Numu/non_osc/np_array_resume/'
b8__osc_numu_dir = 'b8Numu/osc/np_array_resume/'

# Load observables
#energy_unosc_nue_true = np.load(main_dir + b8__nonosc_nue_dir + 'parentKE1.npy')
#energy_osc_nue_true = np.load(main_dir + b8__osc_nue_dir + 'parentKE1.npy')
energy_unosc_nue_recons = np.load(main_dir + b8__nonosc_nue_dir + 'energy.npy')
energy_osc_nue_recons = np.load(main_dir + b8__osc_nue_dir + 'energy.npy')

#energy_unosc_numu_true = np.load(main_dir + b8__nonosc_numu_dir + 'parentKE1.npy')
#energy_osc_numu_true = np.load(main_dir + b8__nonosc_numu_dir + 'parentKE1.npy')
energy_unosc_numu_recons = np.load(main_dir + b8__nonosc_numu_dir + 'energy.npy')
energy_osc_numu_recons = np.load(main_dir + b8__osc_numu_dir + 'energy.npy')
```

# Plots

```python
plt.figure(figsize = (8, 5.5))

bins = 30

sn.histplot(energy_unosc_nue_recons, bins = bins, element = 'step', alpha = 0, label = r'$\nu_e$ Unoscillated')
sn.histplot(energy_unosc_numu_recons, bins = bins, element = 'step', alpha = 0, label = r'$\nu_\mu$ Unoscillated')
sn.histplot(energy_osc_nue_recons, bins = bins, element = 'step', alpha = 0, label = r'$\nu_e$ Oscillated')
sn.histplot(energy_osc_numu_recons, bins = bins, element = 'step', alpha = 0, label = r'$\nu_\mu$ Oscillated')

plt.yscale('log')
plt.legend(loc = 'best')

plt.xlabel('Reconstructed Energy (MeV)')

plt.show()
```
