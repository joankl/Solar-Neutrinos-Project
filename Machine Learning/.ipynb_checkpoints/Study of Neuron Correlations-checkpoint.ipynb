```python
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
from matplotlib import cm
import seaborn as sn

import pandas as pd

import tensorflow as tf

from sklearn.utils import shuffle
```

```python
#Funcion para rellenar con ceros arrays que no producen matrices cuadradas

def reshape_to_square(arr):
    
    n = int(np.ceil(np.sqrt(len(arr))))  # Redondear hacia arriba
    padded = np.pad(arr, (0, n**2 - len(arr)))  # Rellenar con ceros
    return padded.reshape(n, n)
```

## 2.2 - Sum of the |Weights| of Input Layers and 1st Hidden Layer

Lets sum the |weights| of within each row. We should get an array with 100 entries. Then pass to an 10x10 image and comparte with the high-statistical $\mathcal{L}(\Delta,\alpha)$ 10x10. Lets add the value of the energy and radius weights as a new column of the sum_imput_weight

```python
dim_corr = np.ones(10)
dim_corr
```

```python
save = True
image_dim = 10

# Compute sum -------------------------
dim_corr = np.ones(image_dim)  #dimensional correction for the energy and radius. It allows to add the energy and radius weights to the pixel weights

pixel_w_sum = np.sum(np.abs(weights_ly_0), axis = 1)# Sum the weights that the ith input neutron has with the  neutrons of the 1st hidden layer, i.e. sum values of the same row of pixel W matrix
pixel_w_sum = pixel_w_sum.reshape((image_dim,image_dim))

energy_w_sum, radius_sum = np.sum(np.abs(energy_w))*dim_corr, np.sum(np.abs(radius_w))*dim_corr
energy_w_sum, radius_sum = energy_w_sum.reshape(1,image_dim).T, radius_sum.reshape(1,image_dim).T

print(pixel_w_sum.shape)
weights_sum = np.concatenate((pixel_w_sum, energy_w_sum), axis = 1)
weights_sum = np.append(weights_sum, radius_sum, axis = 1)

#Normalize Data to the same scale ----
norm_weights_sum = weights_sum/(np.max(weights_sum))
norm_L_image = L_image/(np.max(L_image))

# Plot -------------------------------
fig, axes = plt.subplots(1, 2,  figsize=(22, 8), gridspec_kw={'wspace': 0.25})

axes[0].imshow(norm_weights_sum, cmap = cm.PuBuGn)
ax1 = axes[1].imshow(norm_L_image, cmap = cm.PuBuGn, extent = [-4, 12, 1, -1], aspect = 'auto')

axes[0].set_title(r'$\Sigma|\text{weights-input-neurons}|$, Energy and Radius', fontsize = 14)
axes[1].set_title(r'High Statistical $\mathcal{L}(\Delta,\alpha)$ - Solar 8B-$\nu_e$', fontsize = 14)

# x-axis and y-axis label of figure weights
pixel_labels = [f"pixel_{i}" for i in range(image_dim)]
en_pos_lables = ['Energy', 'Position']
axes[0].set_xticks(ticks=np.arange(image_dim + 2), labels=pixel_labels + en_pos_lables, rotation=90)
axes[0].set_yticks(ticks=np.arange(image_dim), labels=pixel_labels)

axes[1].set_xlabel(r'$\Delta$ - Time Residual (ns)', fontsize = 12)
axes[1].set_ylabel(r'cos($\alpha$)', fontsize = 12)

cbar = fig.colorbar(ax1, ax = axes, orientation="vertical", fraction=0.05, pad=0.02)

if save:
    plt.savefig('figs/ML_study_weight_mod_sum_vs_solar_image.png', format='png', bbox_inches = 'tight')

plt.show()
```
