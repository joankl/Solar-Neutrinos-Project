```python
import numpy as np
import pandas as pd
import glob
```

# Load Files

## Load the livetime list

```python
main_dir = '/content/drive/MyDrive/SNO+ Analysis/data/solar_data/real_data/bis_MSB/run_lifetime_lists/*'
flist = glob.glob(main_dir)

runID_list, time1_lt, time2_lt = [], [], []

for fdir in flist:

  with open(fdir, "r", encoding="utf-8") as f:
      for linea in f:
          if linea.strip():  # Evitar líneas vacías
              valores = linea.split()  # Separa por espacios o tabulaciones
              runID_list.append(int(valores[0]))
              time1_lt.append(float(valores[1]))
              time2_lt.append(float(valores[2]))

runID_list = np.array(runID_list)
time1_lt = np.array(time1_lt) # RAW LIVETIME (DAYS)
time2_lt = np.array(time2_lt) # NET LIVETIME (DAYS)

# RUN SELECTION:
#runID_selection = (runID_lt >= 300700)

#runID_lt = runID_lt[runID_selection]
#time1_lt = np.array(time1_lt)[runID_selection] # RAW LIVETIME (DAYS)
#time2_lt = np.array(time2_lt)[runID_selection] # NET LIVETIME (DAYS)
```

## Load the filtered runIDs

```python
main_dir = '/content/drive/MyDrive/SNO+ Analysis/data/solar_data/real_data/bis_MSB/Analysis*/filtered_solar_Analysis*.xlsx'
flits = glob.glob(main_dir)

df_list = []

# Load the Excel folders as pd.Dataframe and then concatenate  them as one file
for f in flits:
    df_list.append(pd.read_excel(f))

df = pd.concat(df_list, ignore_index=True)

#Extract the runID from df
runID_data = np.array(df['runID'])
runID_data  = np.unique(runID_data)  #Avoid double counting

#Extract the min and max runID

min_runID_data, max_runID_data = np.min(runID_data), np.max(runID_data)

print(f'min. runID = {min_runID_data}')
print(f'max. runID = {max_runID_data}')
```

# Compute the Dataset Livetime

```python
# Extract the net lifetime (time2_lt) where runID from data is in runID_list
same_runID_condition = np.isin(runID_list, runID_data)

filtered_lt = time2_lt[same_runID_condition]
tot_dataset_lt_days = np.sum(filtered_lt)
print(f'dataset with livetime = {tot_dataset_lt_days} days')
```
