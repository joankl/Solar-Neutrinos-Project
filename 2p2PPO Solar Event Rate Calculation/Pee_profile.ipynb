# Notebook Dedicated to Obtain the Plot of the Survival Probability (Pee) as a Function of Neutrino's Energy

```python
import numpy as np
import pickle

import matplotlib.pyplot as plt
from matplotlib.ticker import MultipleLocator, ScalarFormatter
```

# Load Data

```python
Pee_dir = '/home/joankl/data/solars/mc/PSelmaa/'
Pee_fname_pattern = 'pselmaa_test_sun_pee_'

reaction = 'b8'
ssm = 'B16_GS98_'

PSelmaa_data = np.loadtxt(Pee_dir + Pee_fname_pattern + ssm + reaction + '.txt', skiprows=1)
Pee = PSelmaa_data[:,1]
Pee_energy = PSelmaa_data[:,0]
```

# Plot

```python
fig, ax = plt.subplots(figsize = (7,5.5))
font_style_title = {'family':'serif', 'weight': 'normal','color':'black','size':13}
font_style_axis= {'family':'serif', 'weight': 'normal','color':'black','size':12}

plt.plot(Pee_energy, Pee, linewidth = 1.5, color = 'red')

plt.xlabel('Neutrino Energy (MeV)', fontdict = font_style_axis)
plt.ylabel(r'$P_{ee}$', fontdict = font_style_axis)

plt.title(r'$P_{ee}(E_\nu)$ for GS98 SSM', y = 1.03, fontdict = font_style_title)

ax.xaxis.set_minor_locator(MultipleLocator(0.5))
ax.xaxis.set_major_locator(MultipleLocator(2))

ax.yaxis.set_minor_locator(MultipleLocator(0.01))
ax.yaxis.set_major_locator(MultipleLocator(0.05))

ax.tick_params(which='minor', top=True, bottom=True, left=True, right=True)
ax.tick_params(which='major', top=True, bottom=True, left=True, right=True)

plt.xlim(min(Pee_energy), max(Pee_energy))

plt.savefig('Pee_GS98.png', dpi=300, bbox_inches='tight')

plt.show()
```
